## Data Preprocessing

```python
X = df[['Hours']]
```

```python
y = df[['Scores']]
```

## Exploratory Data Analysis

```python
plt.scatter(X, y, color = 'b')
```

## Model Seletion 

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.1, random_state=42, shuffle = False)
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
X_test.shape
```

```python
y_train.head()
```

## Model Training 

```python
from sklearn.linear_model import LinearRegression
```

```python
slr = LinearRegression()
```

```python
model = slr.fit(X_train, y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
y_predict=model.predict(X_test)
```

```python
y_predict
```

```python
plt.scatter(X, y, color = 'blue')
plt.plot(X_test, y_predict)
```

##  Model Evaluation 

```python
from sklearn.metrics import r2_score
```

```python
from sklearn.metrics import r2_score
R_square=r2_score(y_test, y_predict)
```

```python
R_square
```

```python
from sklearn import metrics
```

```python
Mean_absolute_error=print('mean s error',metrics.mean_squared_error(y_test,y_predict)) 
```
